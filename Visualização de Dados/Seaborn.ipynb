# Visualizando Relações Estatísticas nos Dados.


> A análise estatística é o processo de entender como variáveis no conjunto de dade se relacionam uma com a outra e como essas relações dependem de outras variáveis. A Visualização de dados se torna uma peça essencial nesse processo, pois, quando os dados são visualizados adequadamente, as pessoas conseguem descobrir padrões que indicam relações nos dados.

## Função relplot()

Método de alto nível para visualização de relaçoes estatísticas entre variáveis. O método relplot() já combina o **scatter plot** e **line plots**. 

Muito útil para ver relações entre duas ou mais variáveis com o uso de parâmetros de semântica dos dados.

# Relação de Variáveis com Scatter Plots

**Visualizando a relação entre a variável valor total da conta e gorjeta.**

```python
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', data=tips)
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)**

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável sex**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips);
```

**Visualizando a relação entre as variávels total da conta e gorjeta e colorindo cada ponto com a variável  sexo e definindo um estilo também baseado na variável sexo**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style='sex', data=tips);
```

**É possível representar quatro variáveis alterando a tonalidade e estilo de cada ponto. Mas é preciso cautela, pois, a visualização pode ficar confusa.**

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style='smoker', data=tips);
```

**Em casos de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma seguencial.**

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

**É possível customizar a paleta de cores sequenciais no link**: 

* https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette

**Usando a opção size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados**

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", size="size", hue='size' ,sizes=(15,300), data=tips);
```

## Cores no Seaborn

* A funcao que controla as cores no seaborn e a color_palette(). 
* Essa funcao fornece uma interface para gerar cores no seaborn.
* Usamos esse metodo internamente ao usar o parametro palette.
* O metodo color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib
* O metodo set_palette() recebe os mesmos parametros e eh usado para definir um estilo padrao para todos os plots.
* As cores devem ser definidas levando em consideracao o tipo de dado que queremos plotar. (qualitativos, sequenciais,diverging)

​


*Ferramenta Color Brewer 2.0 *
http://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3

**Paleta de cores para dados Quantitativos**

```python
sns.palplot(sns.color_palette('Accent'))
```

**Paleta de cores para dados Divergentes**

```python
sns.palplot(sns.color_palette('PuOr'))
```

**Paleta de cores para dados Sequênciais**

```python
sns.palplot(sns.color_palette('Blues'))
```

**Gráfico usando a palette de ce cor Sequêncial 'Blues'**

```python
sns.relplot(x='total_bill', y='tip', size='size', hue='size', palette='Greens', sizes=(12,250), data=tips)
```

**Gráfico usando a paleta de cor qualitativa "Set2"**

```python
sns.catplot(x='sex', kind='count', palette='Set1' ,data=tips)
```

**Gráfico usando a palette de cor divergente 'PiYG'**

```python
sns.heatmap(tips.corr(), annot=True, cmap='RdGy')
```

**Definindo cores padrão para gráficos do Seaborn**

```python
sns.set_palette("pastel")

# Usando a paleta de cor pastel
sns.relplot(x='total_bill', y='tip', data=tips)
```

**Instrução With para definir temporariamente a paleta de cor**

```python
with sns.color_palette("Accent"):
    sns.relplot(x='total_bill', y='tip', size='size', sizes=(12,250), data=tips)
```

## Plotando dados categóricos

* Quando trabalhamos com variáveis categóricas temos que visualizar dados de formas diferentes.
* O Seaborn fornece o método catplot() que já contém diversos tipos de gráficos embutidos.
* Isso facilita, pois, você pode usar diferentes gráficos usando um mesmo método.

**Gráficos de barras ou também conhecidos como gráficos de colunas**

```python
sns.catplot(x='sex', kind='count', palette='Set2', data=tips)
```

**Gráficos de barras horizontais usando a coluna Day**

```python
sns.catplot(y='day', kind='count', palette='Set1', data=tips)
```

**Scatter plot com dados categóricos**

```python
with sns.color_palette('Set1'):
    sns.catplot(y='total_bill', x='day', data=tips)
```

**O parâmetro swarm evita a sobreposição de ponto**

```python
sns.catplot(y='total_bill', x='day', kind='swarm' palette='Set1', kind='swarm', data=tips)
```

**O parâmetro hue permite adicionar uma terceira variável a nossa visualização**

```python
with sns.color_palette('Dark2'):
    sns.catplot(y='total_bill', x='day', kind='swarm', hue='sex', data=tips)
```

* O parametro **order** permite alterarmos a ordem padrão das categorias que estão sendo exibidas.

* Isso é útil quando temos mais um gráfico na mesma figura e queremos manter as mesmas ordens.

```python
sns.catplot(y='total_bill', x='day', kind='swarm', hue='sex', palette='Dark2', order=['Sat', 'Sun', 'Fri', 'Thur'], data=tips)
```

**Invertendo a visualização para plots horizontais**

```python
sns.catplot(x='total_bill', y='day', kind='swarm', hue='sex', palette='Dark2', data=tips)
```

## Gráficos com Regressão

* Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
* Podemos visualizar essas informações com linhas de regressão.
* Com modelos de regressão simples é possível checar se existe alguma **correlacão** entre algumas variáveis.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

**Exibe uma linha de regressão para visualizar correlação entre as variáveis**

```python
%matplotlib inline
sns.lmplot(x='total_bill', y='tip', hue='sex', palette='Dark2', data=tips)
```

**Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica *time***

```python
sns.lmplot(y='tip', x='total_bill', hue='smoker', col='time', palette='Pastel1', data=tips)
```

**Parâmetros scatter_kws e line_kws**

* Scatter_kws:
    's' -> Tamanho dos pontos.<br>
    'alpha' -> Opacidade dos pontos.<br>
* Line_kws:
    'lw': Tamanho da reta.

```python
%matplotlib inline
sns.lmplot(y='total_bill',
           x='tip',
           hue='smoker',
           palette='Dark2',
           col='time',
           scatter_kws={'s': 90, 'alpha':0.7},
           line_kws={'lw': 4},
           data=tips)
```


**Segregando gráficos pelo parâmetro *col* e pelo parâmetro *row***

```python
%matplotlib inline
with sns.color_palette('Dark2'):
    sns.lmplot(y='total_bill', 
               x='tip', 
               hue='smoker', 
               col='time', 
               row='sex', 
               scatter_kws={'s':90, 'alpha':0.7}, 
               line_kws={'lw': 3},
               data=tips)
```

## Visualizando distribuição de observações

* Um dataset muito grande será difícil de visualizar variáveis categóricas.
* É preciso sumarizar a distribuição dos dados para facilitar a visualização.
* Podemos usar um gráfico do tipo Boxplot para visualizar a distribuição de tais variáveis.
* Esse tipo de gráfico é muito útil para visualizar ***Outliers***.

**Boxplot de Dias de Semana por Total de Conta**

```python
sns.catplot(x='day', y='total_bill', kind='box', palette='Dark2', data=tips)
```

**Gráfico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não.**

```python
sns.catplot(y='total_bill', x='day', hue='smoker', kind='box', data=tips)
```

* Tipo de Boxplot com foco maior na forma dos dados do que nos outliers.
* Esse tipo de gráfico é interessante quando temos grandes datasets.

```python
sns.catplot(y='total_bill', x='day', kind='boxen', data=tips)
```

* O Método stripplot permite plotar a distribuição dos dados.
* Podemos combinar os dois gráficos para ter mais informação.

```python
sns.catplot(y='total_bill', x='day', kind='boxen', data=tips)
sns.stripplot(y='total_bill', x='day', color='blue', data=tips)
```

**O método catplot() permite usarmos facetgrid assim podemos combinar gráficos em uma única figura**

```python
%matplotlib inline
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='boxen', height=4)
```

## Violin Plot

* Gráficos que ideal para visualizar a distribuição de variáveis.
* Combinação do Boxplot com o KDE.
* Funciona bem para distribuições com picos.
* Permite uma visualização mais rica do que com o Boxplot normal.

**Violin Plot**

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

**Violin Plot Horizontal**

```python
sns.catplot(x='total_bill', y='day', kind='violin', data=tips)
```

**Usando o parâmetro Hue para ter uma terceira variável**

```python
sns.catplot(x='day', y='total_bill', hue='time', kind='violin', data=tips)
```

* **É possível dividir o violin plot com o parâmetro split**
* **Isso é útil temos muitos dados para exibir**

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', split=True, data=tips)
```

* **Usando o parametro inner para preencher o violin plot com observações**
* **Útil para destacar bem as partes que possuem mais dados**

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', split=True, inner='stick', palette='pastel', data=tips)
```

* **Podemos combinar também um gráfico de swarm com o Violin Plot**
* **Dessa forma fica mais explicito ainda a frequencia de pontos nas partes do "violino"**


```python
sns.catplot(x='day', y='total_bill', kind='violin', palette='pastel', data=tips)
sns.swarmplot(x='day', y='total_bill',  color='k', size=3, data=tips)
```

## Joinplot (ScatterPlot e Histograma)

* Quando queremos ver a distribuição de duas variáveis podemos usar o jointplot.
* União de gráficos do tipo Scatter Plot e Histogramas para as duas variáveis.
* Mostra o Scatter plot para as duas variáveis e um histograma para cada variável separadamente.

```python
sns.jointplot(x='tip', y='total_bill', color='r', data=tips)
```

## Hexbin Plots

* Tende a funcionar melhor com as cores em branco de fundo

```python
sns.jointplot(x='tip', y='total_bill', color='r', kind='hex', data=tips)
```

**Usando o KDE para Plotar mais de uma variável**

```python
sns.jointplot(x='tip', y='total_bill', color='b', kind='kde', data=tips)
```

```python
correlacoes
```

```python
ax = sns.heatmap(correlacoes)
```
