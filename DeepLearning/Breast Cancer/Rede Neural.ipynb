```python
import pandas as pd
```

```python
dataset_input = pd.read_csv('entradas_breast.csv')
```

```python
dataset_outputs = pd.read_csv('saidas_breast.csv')
```

```python
dataset_outputs.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_input, dataset_outputs, test_size=0.3)
```

```python
from tensorflow import keras
```

```python
# Criando a rede
from tensorflow.keras.models import Sequential
# Rede full conectada 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
# Units -> Número de Neurônios Quantidade de entradas + um neuronio por ser um problema de classificao binária / 2
# Activation -> Função de ativação 
# kernel_initializer -> pesos da rede Colocando valores aleatórios
# input_dim -> primeira camada
model.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform', input_dim=30))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# opmizer -> Otimiza a correção dos pesos ex: Back propagation, Adam
# loss -> Erro a ser utilizada, por ser uma classificação vamos utilizar o binary_crossentropy
# metrics -> Metrica para acompanhar o desenvolvimento da rede
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
# Batch_size -> A cada 10 amostra atualiza os pesos
# Aprendizado offline = Apresento todas as amostras para a rede ai sim calculo o erro
# Aprendizado online = Calculo do erro a cada amostra
# Misto -> a cada 10 amostra calculo o erro BATCH_SIZE
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
# predizendo os dados
y_predicted = model.predict(X_test)

# Se for maior que 0.5 recebe 1 (True)
y_predicted = (y_predicted > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
performance_output = accuracy_score(y_test, y_predicted)
```

```python
matriz = confusion_matrix(y_test, y_predicted)
```

```python
# Avalização pelo Keras
result = model.evaluate(X_test, y_predicted)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.model_selection import cross_val_score
from tensorflow.keras.layers import Dropout 
from sklearn.model_selection import GridSearchCV
```

```python
def build_model(optimizer, loss, kernel_initializer, activation, neurons):
    model = Sequential()
    model.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer, input_dim=30))
    model.add(Dropout(0.2)) # Para não realizar Overfitting zera 20% caracterisca
    model.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer)) # 2 Camada
    model.add(Dropout(0.2)) # Para não realizar Overfitting
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss=loss, metrics=['binary_accuracy'])
    return model

model = KerasClassifier(build_fn=build_model)

parametros = {'batch_size': [5, 30],
              'epochs': [50, 100],
              'optimizer': ['adam', 'sgd'],
              'loss':['binary_crossentropy', 'hinge'],
              'kernel_initializer': ['random_uniform', 'normal'],
              'activation': ['relu', 'tanh'],
              'neurons': [16,8]
             }

# Criar o grid search
grid_search = GridSearchCV(estimator=model, 
                           param_grid=parametros,
                           scoring='accuracy',
                           cv=5)

grid_search = grid_search.fit(dataset_input, dataset_outputs)
```

```python
resultados = cross_val_score(estimator=model,
                            X=dataset_input, y=dataset_outputs,
                            cv=4, scoring='accuracy')
```

```python
media = resultados.mean()
desvio = resultados.std()
```

```python
# Salvar a rede em disco
model = Sequential()
model.add(Dense(units=8, activation='relu', kernel_initializer='normal', input_dim=30))
model.add(Dropout(0.2)) # Para não realizar Overfitting zera seis caracterisca
model.add(Dense(units=8, activation='relu', kernel_initializer='normal')) # 2 Camada
model.add(Dropout(0.2)) # Para não realizar Overfitting
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
X_train, X_test, y_train, y_teste = train_test_split(dataset_input, dataset_outputs, test_size=0.3)
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
model_json = model.to_json()
```

```python
with open('neural-beast-aula.json', 'w') as json_file:
    json_file.write(model_json)
```
