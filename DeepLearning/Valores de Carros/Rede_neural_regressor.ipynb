```python
valores = {'vehicleType':'limousine',
           'gearbox':'manuell',
           'model':'golf',
           'fuelType':'benzin',
           'notRepairedDamage':'nein'}
df.fillna(value=valores, inplace=True)
```

```python
# TRansformando os dados categoricos em numericos
x = df
x = pd.concat([x, pd.get_dummies(df['abtest'], prefix='abtest')], axis=1)
x.drop('abtest', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['vehicleType'], prefix='vehicleType')], axis=1)
x.drop('vehicleType', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['gearbox'], prefix='gearbox')], axis=1)
x.drop('gearbox', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['model'], prefix='model')], axis=1)
x.drop('model', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['fuelType'], prefix='fuelType')], axis=1)
x.drop('fuelType', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['brand'], prefix='brand')], axis=1)
x.drop('brand', axis=1, inplace=True)
```

```python
x = pd.concat([x, pd.get_dummies(df['notRepairedDamage'], prefix='notRepairedDamage')], axis=1)
x.drop('notRepairedDamage', axis=1, inplace=True)
```

```python
x
```

```python
previsores = x.iloc[:,1:318].values
preco_real = x.iloc[:,0].values
```

```python
regressor = Sequential()
regressor.add(Dense(units=159, activation='relu', input_dim=316))
regressor.add(Dense(units=159, activation='relu'))
regressor.add(Dense(units=1, activation='linear')) # Regressão função linear por padrao a última camada sem o parâm. activation é linear
regressor.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
# Treinar a rede
regressor.fit(previsores, preco_real, batch_size=200, epochs=100)
```

```python
previsoes = regressor.predict(previsores)
```

```python
previsores.mean()
```

```python
preco_real.mean()
```

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
```

```python
model = KerasRegressor(regressor)
```

```python
parametros = {'units':[159, 79]}
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=parametros,cv=5,scoring='neg_mean_absolute_error')
```

```python
grid_search.fit(previsoes, preco_real, epochs=5)
```
